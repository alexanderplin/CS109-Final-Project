```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
aapl_data = pd.read_csv('AAPL.csv', parse_dates = [0], keep_date_col = True)
aapl_data.head(n=5)
```

```python
def filter_data_by_date(data, date_start, date_end):
    return data[(data['date'] > date_start) & (data['date'] < date_end)]

def plot_stock_data(data, ax, option="close", title=""):
    ax.set_title(title)
    ax.plot(data['date'].values, data[option].values)
    return ax

fig, ax = plt.subplots(1, 1, figsize=(16, 4))
plot_stock_data(aapl_data, ax, title="10 years AAPL")
```
